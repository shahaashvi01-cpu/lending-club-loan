# Data Cleaning

```python
# Creating a function to remove columns with more than 50% missing values to improve data quality.
def drop_high_null_columns(df, threshold=0.5):
    # Calculate percentage of missing values
    null_percent = df.isnull().mean()
    
    # Select columns where missing % > threshold
    cols_to_drop = null_percent[null_percent > threshold].index
    
    # Drop those columns
    df = df.drop(columns=cols_to_drop)
    
    return df
```

```python
df = drop_high_null_columns(df)

print(df.shape)
```

```python
df.head()
```

```python
# Removing single values unique columns 
def drop_single_value_columns(df):
    # Find columns with only one unique value
    cols_to_drop = df.columns[df.nunique() <= 1]
    
    # Drop those columns
    df = df.drop(columns=cols_to_drop)
    
    return df
```

```python
def case_function(df, cols, lower=True):
    df = df.copy()
    for col in cols:
        if lower:
            df[col] = df[col].str.strip().str.lower()
        else:
            df[col] = df[col].str.strip().str.upper()
    return df
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style='whitegrid')
plt.figure(figsize=(8,6), dpi=100)

# Step 1: create income bins
def income_bin(x):
    if x <= 45000:
        return 'Low'
    elif x <= 90000:
        return 'Medium'
    elif x <= 150000:
        return 'High'
    else:
        return 'Very High'

df['annual_inc_bin'] = df['annual_inc'].apply(income_bin)

# Step 2: group data
temp = df.groupby(['annual_inc_bin', 'loan_status'], observed=False).size().reset_index(name='count')

# Step 3: percentage within each income group
temp['pct'] = temp.groupby('annual_inc_bin')['count'].transform(lambda x: x / x.sum() * 100)

# Step 4: plot
ax = sns.barplot(
    x='annual_inc_bin',
    y='pct',
    hue='loan_status',
    data=temp,
    palette='Paired'
)

plt.xlabel('Income Category')
plt.ylabel('Percentage (%)')
plt.title('Defaulter Rate by Income Category')
plt.ylim(0, 100)

# Step 5: add values
for p in ax.patches:
    ax.text(
        p.get_x() + p.get_width()/2,
        p.get_height(),
        f'{p.get_height():.1f}%',
        ha='center',
        va='bottom',
        fontsize=9
    )

plt.legend(title='Loan Status')
plt.tight_layout()
plt.show()
```

The Low income category has the highest default rate at 17.4%, while the High and Very High income categories have lower default rates of around 10.9% and 11.0% respectively.
Non-defaulters dominate all income groups, but the graph indicates that borrowers with lower income are more likely to default compared to higher-income borrowers.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style='whitegrid')
plt.figure(figsize=(12,6), dpi=100)

# Step 1: filter defaulters
df_def = df[df['loan_status'] == 1]

# Step 2: percentage by state
temp = (
    df_def['addr_state']
    .value_counts(normalize=True)
    .mul(100)
    .reset_index()
)
temp.columns = ['state', 'pct']

# Step 3: plot (FIX warning using hue)
ax = sns.barplot(
    x='state',
    y='pct',
    data=temp,
    hue='state',
    palette='Reds',
    legend=False
)

plt.xlabel('State')
plt.ylabel('Defaulter (%)')
plt.title('State-wise Defaulter Distribution')
plt.xticks(rotation=45)
plt.ylim(0, temp['pct'].max() + 2)

# Step 4: add values (tilted)
for p in ax.patches:
    ax.text(
        p.get_x() + p.get_width()/2,
        p.get_height(),
        f'{p.get_height():.1f}%',
        ha='center',
        va='bottom',
        rotation=45,
        fontsize=9
    )

plt.tight_layout()
plt.show()
```

California (CA) has the highest share of defaulters at 20.0%, followed by Florida (FL) at 9.0% and New York (NY) at 8.8%, indicating these states contribute the most to total loan defaults.
Most other states have comparatively low default percentages below 5%, while states like NE, MS, TN, and ID show almost negligible default contribution.
